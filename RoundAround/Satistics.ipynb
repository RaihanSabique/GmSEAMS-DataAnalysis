```python
from datetime import date
from datetime import datetime
def calculate_age(born):
    if(not isinstance(born, str)):
        return int(born)
    born = datetime.strptime(born, '%Y-%m-%d')
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
```

```python
import math
from datetime import date
from datetime import datetime
def time_diff(x, y):
    # print(type(x), type(y))
    # a = datetime.strptime(x, '%Y-%m-%d %H:%M:%S.%f')
    # b = datetime.strptime(y, '%Y-%m-%d %H:%M:%S.%f')
    return (y - x).total_seconds() # in seconds

angle_prev = 0
def calculate_phase_diff(right_df, left_df):
    # print(right_df, left_df)
    right_df['Timestamp'] = pd.to_datetime(right_df['righttap_time'])
    left_df['Timestamp'] = pd.to_datetime(left_df['lefttap_time'])
    
    df = pd.merge_asof(right_df, left_df, on='Timestamp')
    # for index, row in df.iterrows():
    #     print(row["righttap_angle_sin"], row["lefttap_angle_sin"], math.asin(row["righttap_angle_sin"])*( 180.0 / math.pi ), math.asin(row["lefttap_angle_sin"])*( 180.0 / math.pi ))
    df["phase_diff"] = df.apply(lambda x: abs(math.asin(x.righttap_angle_sin)*( 180.0 / math.pi )- math.asin(x.lefttap_angle_sin)*( 180.0 / math.pi )), axis=1)
    # df["speed_diff"] = df.apply(lambda x: x.righttap_speed- x.lefttap_speed, axis=1)
    
    # print(df)
    return df[["Timestamp", "phase_diff", "righttap_angle", "righttap_angle_sin", "lefttap_angle", "lefttap_angle_sin"]]
def calculate_angular_velocity_right(df_right):
    df_right["angle"] = df_right.apply(lambda x: math.asin(x.righttap_angle_sin), axis=1)
    angle_diff = df_right.angle.diff()
    time_diff = df_right.time_diff.diff()
    return angle_diff/time_diff

def calculate_angular_velocity_left(df_left):
    df_left["angle"] = df_left.apply(lambda x: math.asin(x.lefttap_angle_sin), axis=1)
    angle_diff = df_left.angle.diff()
    time_diff = df_left.time_diff.diff()
    return angle_diff/time_diff
    
def calculate_speed_diff(right_df, left_df):
    # print(right_df, left_df)
    right_df['Timestamp'] = pd.to_datetime(right_df['righttap_time'])
    left_df['Timestamp'] = pd.to_datetime(left_df['lefttap_time'])
    
    df = pd.merge_asof(right_df, left_df, on='Timestamp')
    
    # df["speed_diff"] = df.apply(lambda x: x.righttap_speed- x.lefttap_speed if(x.righttap_speed>x.lefttap_speed) else x.lefttap_speed-x.righttap_speed, axis=1)
    # print(df)
    df["speed_diff"] = df.apply(lambda x: x.righttap_speed- x.lefttap_speed, axis=1)
    return df[["Timestamp", "speed_diff", "righttap_speed", "lefttap_speed"]]
radius = 199.16
def calculate_angular_displacement(df):
    # print(right_df, left_df)
    df["sin_inverse_right"] = df.apply(lambda x: math.asin(x.righttap_angle_sin)*radius, axis=1)
    df["sin_inverse_left"] = df.apply(lambda x: math.asin(x.lefttap_angle_sin)*radius, axis=1)
    df["path_right"] = df['sin_inverse_right'].diff().abs()
    df["path_left"] = df['sin_inverse_left'].diff().abs()
    return df["path_right"].sum(), df["path_left"].sum()
```

```python
import numpy as np
import scipy.stats
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    # print(a)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h
```
